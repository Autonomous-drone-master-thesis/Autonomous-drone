```python
import requests
```

```python
models_url = "https://raw.githubusercontent.com/tensorflow/models/master/research/object_detection/g3doc/tf2_detection_zoo.md"
```

```python
def get_md_file_from_github(url):
    response = requests.get(url)

    if response.status_code == 200:
        return response.text
    else:
        raise Exception(f"Error fetching the file: {response.status_code} - {response.reason}o")
```

```python
data = get_md_file_from_github(models_url)
```

```python
import re
```

```python
models = []
pattern = re.compile(r'(.+?)\s+(\d+x\d+)')
for line in data[data.find('Model name'):].splitlines():
    if line.startswith('['):
        model_name_with_size = line[1:line.find(']')]
        
        match = pattern.match(model_name_with_size)
        if match:
            model_name, size = match.groups()

            download_link = line[line.find('(')+1:line.find(')')]
            
            speed, coco_map, output = line.split('|')[1:]
                

            model_dict = {
                "model": model_name,
                "size": size,
                "download_link": download_link,
                "speed": speed.strip(),
                "coco_map": coco_map.strip(),
                "output": output.strip(),
                "downloaded": False
            }

            models.append(model_dict)
```
