```python
import requests
from bs4 import BeautifulSoup
```

```python
models_url = "https://raw.githubusercontent.com/tensorflow/models/master/research/object_detection/g3doc/tf2_detection_zoo.md"
```

```python
def get_md_file_from_github(url):
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        code_section = soup.find("pre")
        print(soup.find("body"))
        if code_section:
            return code_section.text
        else:
            raise Exception("Error: Unable to find Markdown content in the provided URL.")
    else:
        raise Exception(f"Error fetching the file: {response.status_code} - {response.reason}")
```

```python
get_md_file_from_github(models_url)
```
